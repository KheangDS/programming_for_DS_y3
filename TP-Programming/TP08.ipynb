```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Q1

```python
df = pd.read_csv(r'C:\Programming for DS Y3-S2\TP-Programming\company_sales_data.csv')
print(df.shape)
df.head(8)
```

## Q2

```python
df.info()
```

## Q3

```python
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(df["total_profit"])
plt.title("Company Profit Per Month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
```

## Q4

```python
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(df["total_profit"],c='red',linestyle='--',marker='o',markersize=10,linewidth=5,markerfacecolor='black')
ax.legend(['Profit Data of Last year'],loc='lower right')
plt.title("Company Sale Data of Last Year")
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
```

## Q5

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(df['month_number'], df['facecream'],marker='o', label='Face Cream')
ax.plot(df['month_number'], df['facewash'],marker='o', label='Face Wash')
ax.plot(df['month_number'], df['toothpaste'],marker='o', label='Toothpaste')
ax.plot(df['month_number'], df['bathingsoap'],marker='o', label='Bathing Soap')
ax.plot(df['month_number'], df['shampoo'],marker='o', label='Shampoo')
ax.plot(df['month_number'], df['moisturizer'],marker='o', label='Moisturizer')

plt.title('Sales Data ')
plt.xlabel('Month Number')
plt.ylabel('Sales in Units number')
plt.legend()
plt.show()
```

## Q6

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['month_number'], df['toothpaste'], color='blue', marker='o',label='Toothpaste Sales')
plt.title('Toothpaste Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.grid(True, linestyle='--', linewidth=0.7)
plt.legend()
plt.show()
```

plt.legend() displays a legend on the plot, which helps identify each plotted data series by its label.
When you add a label argument to your plot commands (like ax.plot(..., label='Face Cream')), plt.legend() will show those labels in a box on the graph, making it easier to distinguish between different lines or markers.

## Q7

```python
import numpy as np

bar_width = 0.4 #sets the width of each bar.
months = df['month_number']
x = np.arange(len(months))

fig, ax = plt.subplots(figsize=(10,6))
ax.bar(x - bar_width/2, df['facecream'], width=bar_width, label='Face Cream')
ax.bar(x + bar_width/2, df['facewash'], width=bar_width, label='Face Wash')

#The first ax.bar plots the face cream sales, shifting the bars left by half the bar width.
# The second ax.bar plots the face wash sales, shifting the bars right by half the bar width.
# This way, bars for each month appear side by side for comparison.

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Face Cream and Face Wash Sales per Month')
plt.grid(True, linestyle='--', linewidth=0.7)
plt.xticks(x, months)
plt.legend()
plt.show()
```

## Q9

```python
plt.figure(figsize=(10,6))
plt.hist(df['total_profit'], bins=5, color='purple', edgecolor='black')
plt.title('Distribution of Total Profit per Month')
plt.legend(['Profit Data'], loc='upper right')
plt.xlabel('Total Profit')
plt.ylabel('Number of Months')
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()
```

## Q10

```python
import plotly.express as px

# List of product columns (excluding 'month_number', 'total_units', 'total_profit')
product_cols = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

# Calculate total units sold per product for the year
total_sales = df[product_cols].sum()

# Create a DataFrame for plotting
sales_df = total_sales.reset_index()
sales_df.columns = ['Product', 'Total_Units']

# Plot pie chart
fig = px.pie(
    sales_df,
    names='Product',
    values='Total_Units',
    title='Total Units Sold per Product (Yearly)',
    hole=0.0
)
fig.update_traces(textinfo='percent+label+value')
fig.show()
```
