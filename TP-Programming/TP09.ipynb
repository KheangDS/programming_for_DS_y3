## Q2

data=data:

This is the DataFrame you're passing in. It should contain numeric columns for plotting and a categorical column (species here) for coloring.

hue='species':

This tells Seaborn to color the data points based on the values in the species column. Each unique value in species will be assigned a different color and marker.

diag_kind='kde':

This sets the type of plot shown on the diagonal subplots (i.e., the plots of each variable against itself). 'kde' means Kernel Density Estimation — a smoothed version of a histogram.

palette='Dark2':

This sets the color palette. 'Dark2' is a built-in Matplotlib color palette that’s good for categorical data.

markers=['o', 's', 'D']:

This defines the marker shapes used for each class in the species column:

'o' = circle

's' = square

'D' = diamond

```python
# The purpose of pairplot() is to plot pairwise relationships in a dataset
sns.pairplot(data=data, hue='species', diag_kind='kde',palette='Dark2',markers=['o','s','D'])
```

1. Which feature pairs best separate the three iris species?
From the scatter plots in the off-diagonal grid:

Petal Length vs Petal Width is the most effective pair for separating the three species:

Setosa is clearly separated from the other two.

Versicolor and Virginica form distinct, though slightly overlapping(very similar), clusters.


Petal Length vs Sepal Length and Petal Width vs Sepal Length are also quite effective at separation, especially between Setosa and the others.


Sepal Length vs Sepal Width offers the least separation, especially between Versicolor and Virginica, which are more intermixed.


✅ Conclusion:
Best feature pairs for species separation:

petal_length vs petal_width

petal_length vs sepal_length



2. Setosa shows distinct, tight KDE peaks, making it easy to classify.

Versicolor and Virginica have overlapping distributions, making them harder to distinguish(to recognize or show the difference) using some features.


3. Using both marker shape and color improves clarity, accessibility, and visual distinction between species, especially when species overlap in feature space.
